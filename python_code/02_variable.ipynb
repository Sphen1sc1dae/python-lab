# 2. 변수 이름을 붙이는 규칙(1)

### 🎯반드시 지켜야 하는 규칙
- 숫자로 시작하지 않는다.
- 공백이 포함되지 않는다.
- 기호가 포함되지 않는다.
- 예약어(키워드)는 사용하지 않는다.

```python
# 잘못된 변수 정의
2var = "파이썬" # 숫자로 시작하여 에러 발생
```

```python
happy python = '파이썬' # 공백 포함으로 에러 발생
```

```python
happy! = '해피' # 기호포함 에러 발생
```

```python
def = "함수" # 예약어 사용으로 에러 발생
```

```python
happyPython = '파이썬' # camelCase 타입으로 정의
happyPython
```

```python
happy_python = '파이썬' # snakeCase 타입으로 정의
happy_python
```

# 3. 변수 이름을 붙이는 규칙(2)

### 🎯 오류는 안나지만 참고할 사항
- 대소문자 구분
- 한 번에 여러개의 변수 선언 가능
- 하나의 값을 여러 변수에 저장 가능

```python
# 1. 대소문자를 구분
abc = 100
abc
```

```python
ABC = 200
ABC
```

```python
print(abc, ABC)
```

```python
# 2. 한 번에 여러개 변수 선언 가능
x, y, z = 100, 200, 300
```

```python
name1, name2 = '홍길동', '이순신'
print(name1, name2)
```

```python
# 3. 하나의 값을 여러 변수에 저장 가능
xx = yy = zz = "Apple"
xx = 100
print(xx, yy, zz)
```
