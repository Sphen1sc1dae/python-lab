# print()
## 1. 기본문법
- print(value, ..., sep = '구분기호', end = '\n', flush = False)

```python
a = 5
b = 5
a + b
```

```python
print(a)
print(b)
print(a + b)
```

```python
print(1, 2, 3, sep = '-', flush = True) # -> flush = True 를 하게 되면 메모리에서도 출력된 값에 대한 내용이 사라진다.
```

```python
print(a, end = '\t')
print(b)
print(a + b)
```

```python
name = '홍길동'
age = 30
print('이름 : ', name, '나이 : ', age)
```

```python
print('이름 : ' + name, '나이 : ' + str(age))
```

```python
# 3. % 포맷 (old version)
print('이름 : %s, 나이 : %d' %(name, age))
```

```python
# 4. format{} 포맷 (new version)
print('이름: {}, 나이: {}' .format(name,age))
```

```python
# 5. f-string (권장)
print(f"이름: {name}, 나이: {age}")
```

```python
# 줄바꿈 없이 옆으로 출력
for i in range(5) :
    print(i, end = ' ')
```
