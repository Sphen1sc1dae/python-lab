# 배열 연산
- 배열의 shape이 같으면 데이터별로 연산이 진행됨

```python
import numpy as np
```

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
print(arr1.shape, arr2.shape)
print(arr1, arr2)
```

```python
print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 / arr2)
```

```python
# shape이 다른 경우 ValueError 발생 💥
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([4, 5, 6])
print(arr1.shape, arr2.shape)
print(arr1 + arr2)
```

- 스칼라(단일 숫자)와 연산

```python
print(arr1 + 10)
print(arr1 - 1)
print(arr1 * 3)
print(arr1 / 2)
```

- 2차원 배열 연산

```python
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])
print(arr1.shape, arr2.shape)
```

```python
print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 / arr2)
```

# 정렬 - sort()와 argsort()
### (1) sort()
- numpy.sort() vs numpy.ndarray.sort() : 원본 배열의 유지 여부가 핵심
- 원본 유지 O : numpy.sort()
- 원본 유지 X : numpy.ndarray.sort()

```python
org_arr = np.array([2, 1, 3, 6, 9])
```

```python
sort_arr = np.sort(org_arr)
sort_arr
```

```python
print(org_arr, sort_arr)
```

```python
# 원본 배열을 직접 오름차순 정렬
org_arr.sort()
print(org_arr)
```

- 내림차순 정렬 : 배열[::-1]

```python
org_arr = np.array([8, 4, 9, 2, 5])
```

```python
sort_arr = np.sort(org_arr)[::-1]
```

```python
print(sort_arr)
```

```python
org_arr.sort()
org_arr[::-1]
print(org_arr)
```

```python
arr = np.array([[8, 12], [7, 1]])
```

```python
sort_arr = np.sort(arr, axis=0) # axis = 0 행 기준으로 정렬
sort_arr
```

```python
sort_arr2 = np.sort(sort_arr, axis=1) # axis = 0 행 기준으로 정렬
sort_arr2
```

### (2) argsort()
- 정렬된 인덱스 반환

```python
names = np.array(['홍길동','김유신','강감찬','이순신', '스미스'])
scores = np.array([72, 81, 95, 66, 88])
```

```python
# 성적순 names 배열 오름차순 정렬
sort_idx = np.argsort(scores)
sort_idx
```

```python
# 정렬한 인덱스를 팬시 인덱싱 실행
names[sort_idx]
```

```python
# 성적순 names 배열 내림차순
sort_idx2 = np.argsort(scores)[::-1]
names[sort_idx2]
```

```python
arr = np.arange(1, 10)
arr
```

```python
print(np.sum(arr))
print(np.mean(arr)) # 평균
print(np.max(arr))
print(np.min(arr))
print(np.std(arr)) # 표준편차
print(np.var(arr)) # 분산
print(np.median(arr)) # 중앙값
```

```python
arr = np.array([2, 3, 1, 20, 7, 8])
print(np.sum(arr))
```

## 2차원 배열에서 집계함수 사용

## (1) 행렬 내적(dot product)

```python
A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([[7, 8], [9, 10], [11, 12]])
C = np.dot(A, B)
print(A)
print(B)
print(C)
```
