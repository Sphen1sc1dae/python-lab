# DataFrame Index

## Index란?
- Pandas의 Index는 데이터프레임과 시리즈의 각 행(row)를 고유하게 식별하는 역할을 수행함
- 관계형 데이터베이스의 Primary key(pk)와 유사한 개념
- 특징
  - 데이터 식별 역할 수행
  - 숫자, 문자열도 가능
  - 생성 후 직접 수정 불가
  - reset_index() 함수를 이용하여 재생성 가능

```python
import warnings
warnings.filterwarnings(action='ignore')

import pandas as pd
import numpy as np
```

## 2. Index 기반 연산 방식
- ✨ 판다스는 인덱스 이름을 기준으로 데이터를 맞춘 후 연산함

### (1) Series 연산
- ✨ 동일한 인덱스를 가진 데이터끼리만 연산됨

####  기본 Series 생성 및 연산

```python
# Series 생성
s1 = pd.Series([10, 20, 30])
s2 = pd.Series([10, 20, 30])
print(s1, '\n', s2)
```

```python
# Series 사칙 연산
print(s1 + s2)
print(s1 - s2)
print(s1 * s2)
print(s1 / s2)
```

#### Index 지정


```python
s1 = pd.Series([10, 20, 30], 
                index=['a', 'b', 'c'])
s2 = pd.Series([10, 20, 30], 
                index=['b', 'c', 'd'])
print(s1, '\n', s2)
```

#### Index 기준 연산 - 동일한 인덱스끼리 연산 진행

```python
# Series 사칙 연산 - 동일한 인덱스 기준으로 연산이 진행됨
print(s1 + s2)
print(s1 - s2)
print(s1 * s2)
print(s1 / s2)
```

### (2) DataFrame 연산
- ✨ 동일한 인덱스, 동일한 컬럼명 기준으로 연산이 진행됨

#### DataFrame 생성

```python
df1 = pd.DataFrame(
    {
        'col1' : [10, 20, 30],
        'col2' : [10, 20, 30]
    },
    index=['a', 'b', 'c']
)
```

```python
df2 = pd.DataFrame(
    {
        'col1' : [10, 20, 30],
        'col3' : [10, 20, 30]
    },
    index=['a', 'b', 'c']
)
```

```python
print(df1)
print()
print(df2)
```

```python
# 인덱스, 컬럼명이 일치해야만 연산이 수행됨
df1 + df2
```

```python
# df1 객체에 col3 컬럼을 추가, 값 100
df1['col3'] = 100
df1
```

```python
df1 + df2
```

## 3. reset_index()
- ✨ reset_index() 실행 시 기존 Index는 첫번째 컬럼으로 이동하고, 새로운 숫자의 Index가 생성됨

```python
# 데이터셋 확인
titanic_df.head(3)
```

```python
# 인덱스 재생성, reset_index()
td_reset_df = titanic_df.reset_index(inplace=False)
td_reset_df.head()
```

```python
# td_reset_df 객체의 인덱스 재생성
td_reset_df2 = td_reset_df.reset_index(inplace=False)
td_reset_df2.head(3)
```

```python
# td_reset_df2 객체의 level_0 컬럼 삭제, 원본 반영
td_reset_df2.drop('level_0', axis=1, inplace=True) 
td_reset_df2.head()
```

```python
# td_reset_df2 객체의 0, 1, 2 인덱스 삭제, 원본 반영
td_reset_df2.drop([0, 1, 2], axis=0, inplace=True)
td_reset_df2.head()
```

```python
# td_reset_df2 객체의 인덱스 재생성, 원본 반영
td_reset_df2.reset_index(inplace=True)
td_reset_df2.head()
```

```python
td_reset_df2.drop(['level_0', 'index'], axis=1, inplace=True)
td_reset_df2
```

## 4. rename() - 컬럼명 변경
- 문법

  DataFrame.rename(columns={'기존 컬럼명' : '신규 컬럼명'})

```python
# Age_0 컬럼 추가, 값은 0
titanic_df['Age_0'] = 0
titanic_df.head(3)
```

```python
# Age_0 컬럼의 데이터를 1로 수정
titanic_df['Age_0'] = 1
titanic_df.head(3)
```

```python
# Age_0 컬럼명을 Age_1 으로 수정
titanic_df.rename(columns={ 'Age_0': 'Age_1'}, inplace=True)
titanic_df.head()
```

```python
# 인덱스 0, 2, 4 => 'a', 'b', 'c' 변경
td_rename = titanic_df.rename(index= {0: "a", 2: "b", 4: "c"})
td_rename.head()
```

```python
td_rename_index = td_rename.index
td_rename_index
```
