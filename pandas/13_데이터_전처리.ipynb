```python
# [예제 1] 결측값 확인 (code7_1, code7_2)
import pandas as pd
import numpy as np

df = pd.read_csv('data/iris.csv')

# 인위적으로 결측값 주입
df.iloc[0, 1] = pd.NA
df.iloc[0, 2] = pd.NA
df.iloc[1, 2] = np.nan
df.iloc[2, 3] = None

print('컬럼별 결측값 수:')
print(df.isnull().sum())
print()
print('결측값이 있는 행:')
print(df.loc[df.isnull().sum(axis=1) > 0, :])
```

```python
# [예제 2] 결측값 제거
import pandas as pd
import numpy as np

df = pd.read_csv('data/iris.csv')
df.iloc[0, 1] = pd.NA
df.iloc[1, 2] = np.nan

print('정제 전 행 수:', len(df))
df_clean = df.dropna().reset_index(drop=True)  # 결측값 행 제거 + 인덱스 재설정
print('정제 후 행 수:', len(df_clean))
print(df_clean.head())
```

```python
# [예제 3] 이상치 탐지 (code7_4)
import pandas as pd
import numpy as np
from scipy import stats

df = pd.read_csv('data/iris.csv')
sw = df['SepalWidthCm']

# Z-score 방법 (|z| > 2 기준)
z = np.abs(stats.zscore(sw))
outliers_z = sw[z > 2]
print('Z-score 이상치 (|z|>2):')
print(outliers_z.values)

# IQR 방법
Q1 = sw.quantile(0.25) #2.800000
Q3 = sw.quantile(0.75) #3.300000
IQR = Q3 - Q1
outliers_iqr = sw[(sw < Q1 - IQR*1.5) | (sw > Q3 + IQR*1.5)]
print(f'\nIQR 이상치 (Q1-1.5×IQR={Q1-1.5*IQR:.2f}, Q3+1.5×IQR={Q3+1.5*IQR:.2f}):')
print(outliers_iqr.values)

# 이상치 제거
clean = sw.loc[~sw.isin(outliers_iqr)]
print(f'\n원본 크기: {len(sw)} → 이상치 제거 후: {len(clean)}')
```

```python
# [예제 4] 정렬 (sort_values), 순위 (rank) (code7_5, code7_6)
import pandas as pd

df = pd.read_csv('data/iris.csv')

# 정렬
print('Sepal_Length 오름차순 상위 5행:')
print(df.sort_values('SepalLengthCm').head())

print('\nSpecies, Sepal_Width 복합 정렬 상위 5행:')
print(df.sort_values(['Species', 'SepalWidthCm']).head())

# 순위
print('\nPetal_Length 오름차순 순위 상위 5:')
print(df['PetalLengthCm'].rank().astype(int).head())
```
