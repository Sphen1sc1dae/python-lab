# DataFrame Sort, Aggregation, GroupBy

```python
import warnings
warnings.filterwarnings(action='ignore')

import pandas as pd
import numpy as np
```

```python
# Titanic 데이셋 로드
titanic_df = pd.read_csv('data/titanic/train.csv')
titanic_df.head()
```

## 1. DataFrame 정렬(Sort)

### (1) sort_values()
- DataFrame을 특정 컬럼 기준으로 정렬할 때 사용함
- 문법

  DataFrame.sort_values(by='컬럼리스트', ascending=True)

  ascending=True(오름차순), ascending=False(내림차순)

```python
# Name 기준 오름차순 정렬
td_sort_name = titanic_df.sort_values(by = 'Name')
td_sort_name.reset_index(inplace=True)
td_sort_name.drop('index', axis=1, inplace=True)
td_sort_name
```

```python
# Pclass 컬럼 기준으로 내림차순 정렬 한 후 td_sort_pclass 객체 생성
td_sort_pclass = titanic_df.sort_values(by='Pclass', ascending=False)
# 인덱스 재생성, 현재 인덱스 컬럼저장 x, td_sort_pclass 객체에 반영
td_sort_pclass.reset_index(drop=True, inplace=True)
td_sort_pclass
```

```python
# 여러 컬럼 정렬
titanic_df.sort_values(by=['Name', 'Pclass']).head()
```

## 2. 집계 함수(Aggregation)

### (1) count()
- 결측치(NaN) 제외
- 실제 데이터 갯수 반환

```python
titanic_df.count()
```

```python
titanic_df['Name'].count()
```

```python
titanic_df[['Name', 'Age']].count()
```

### (2) shape
- 전체 데이터 수를 확인하는 작업에서는 count() 보다 shape 사용이 효율적

```python
titanic_df.shape
```

```python
titanic_df[['Name', 'Age']].shape
```

### (3) mean()
- 평균 계산

```python
titanic_df.info()
```

```python
# 나이 컬럼의 평균 계산
print(titanic_df['Age'].mean()) 
print(int(titanic_df['Age'].mean()))        # 소수점 삭제 => int() 타입 변환, 'ceil, floor 함수는 numpy'
print(titanic_df['Age'].mean().astype(int)) # 소수점 삭제 => int() 타입 변환
print(titanic_df['Age'].mean().round())
```

```python
# 요금의 평균 계산 - 'Fare' 컬럼
print(titanic_df['Fare'].mean()) 
print(int(titanic_df['Fare'].mean()))        # 소수점 삭제 => int() 타입 변환, 'ceil, floor 함수는 numpy'
print(titanic_df['Fare'].mean().astype(int)) # 소수점 삭제 => int() 타입 변환
print(titanic_df['Fare'].mean().round())
```

### (4) sum()
- 합계 계산 함수

```python
titanic_df.sum()
titanic_df.info()
titanic_df[['Age', 'Fare']].sum()
```

```python
# Age, Fare 컬럼을 합계
titanic_df[['Age', 'Fare']].sum().round(2)
```

### (5) min()/max()
- 최소값/최대값

```python
# 수치형 데이터 컬럼을 대상으로 최대, 최소값 조회`
print(titanic_df['Age'].max())
print(titanic_df['Age'].min())
```

## 3. GoupBy
#### groupby()
- 특정 컬럼 기준으로 데이터를 그룹화 할 때 사용
- 문법
  DataFrame.groupby('컬럼리스트')

```python
titanic_df.info()
```

```python
# 객실 등급별 인원수 조회
titanic_df.groupby('Pclass')
```

```python
titanic_df.groupby(['Embarked', 'Pclass'])[['Pclass', 'Fare']].count()
```

```python
# 항구별 그룹, count(), sum()
titanic_df.agg(
    {
        'Pclass' : 'count',
        'Fare' : 'sum'
    }
)
```

```python
# 객실 등급별 그룹, Age : 최대, Fare : 평균
titanic_df.groupby('Pclass').agg(
    {
        'Age' : 'max',
        'Fare' : 'mean'
    }
)
```

```python
# 여러 컬럼에 agg() 함수 적용
titanic_df[['Pclass', 'Fare']].agg(['max', 'min', 'mean'])
```
