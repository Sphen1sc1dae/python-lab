# 결손 데이터(결측치) 처리
### 결손데이터(결측치)란?
- 결측치(Missing Data)는 값이 존재하지 않는 데이터를 의미
- 판다스에서는 NaN으로 표현됨

1) 실무자와 협의한 후 진행 - 제거도 한 방법
2) 평균으로 대체하는 값, 중앙값으로 대체하는 값

```python
import warnings
warnings.filterwarnings(action='ignore')

import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(
    {
        'Name' : ['Smith', 'Lee', 'Kim', 'Park'],
        'Age' : [20, np.nan, 30, 40],
        'Score' : [90, 85, np.nan, 75]
    }
)
df
```

```python
df.describe()
```

```python
df
df.isna()
```

```python
df.isnull()
```

```python
print(df.isna().sum())
print(df.isnull().sum())
```

## 3.결손 데이터 삭제

```python
filter_na = df.dropna()
```

```python
print(df)
```

```python
print(filter_na)
```

```python
filter_na_cols = df.dropna(axis=1)
```

```python
print(filter_na_cols)
```

```python
# 특정 컬럼만 삭제
df.dropna(axis=0, subset=['Age'])
```

```python
df.dropna(how='all', axis=1, inplace=True)
```

```python
df
```

```python
df.dropna(how='any', axis=0, inplace=False)
```
